```python
import networkx as nx
import numpy as np
import pandas as pd
import osmnx as ox
import geopandas as gpd
import requests, fiona
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.colors as colors
%matplotlib inline
ox.config(log_console=True, use_cache=True)
ox.__version__
```

```python
# define place in the following order: 'City, State, Country'
place = 'Rockford, Illinois, USA'
# load the graph based on 'drive' network
G = ox.graph_from_place(place, network_type='drive')
# plot the graph based on place
fig, ax = ox.plot_graph(G, bgcolor = 'white', edge_color = 'gray', node_color = '#2c7fb8',
                        edge_linewidth=0.5, edge_alpha = 0.5, node_size = 1)

with fiona.Env(OSR_WKT_FORMAT="WKT2_2018"):
    # saves nodes, edges as shapefiles into the folder: Rockford
    ox.save_graph_shapefile(G, f"{place.split(',')[0]}")
```

```python
import time
# log start time
start = time.time()
# calculate the closeness centrality using networkx module
# line_graph() method converts nodes to edges and edges to nodes
edge_centrality_closeness = nx.closeness_centrality(nx.line_graph(G))
# log end time
end = time.time()
# calculate the elapsed time for the closeness centrality
print(f'Elapsed time -> {round(float(end-start),2)} seconds')
```

```python
# list of edge values for the orginal graph
ev = [edge_centrality_closeness[edge + (0,)] for edge in G.edges()]

# color scale converted to list of colors for graph edges
norm = colors.Normalize(vmin=min(ev)*0.8, vmax=max(ev))
cmap = cm.ScalarMappable(norm=norm, cmap=cm.hot)
ec = [cmap.to_rgba(cl) for cl in ev]


# color the edges in the original graph with closeness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='k', node_size=0, node_color='w', node_edgecolor='gray', node_zorder=2,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1, filepath=r"images/edge_centrality_closeness.png", 
                        save=True, show=False, close=True, dpi=300)

# color the edges in the original graph with closeness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='k', node_size=0,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1)
```

<b>Closeness centrality</b> quantifies how close is a node to all the other nodes in a network along shortest paths. From the above plot, I could infer that there is a high closeness centrality at the central part of Rockford city. Therefore, there are more nodes connected to other nodes at the center of the network allowing better navigation through movement across the central part of the city.

```python
import time
start = time.time()
# calculate the betweenness centrality using networkx module
edge_centrality_betweenness = nx.betweenness_centrality(nx.line_graph(G))

end = time.time()
print(f'Elapsed time -> {round(float(end-start),2)} seconds')
```

```python
# list of edge values for the orginal graph
ev = [edge_centrality_betweenness[edge + (0,)] for edge in G.edges()]

# color scale converted to list of colors for graph edges
norm = colors.Normalize(vmin=min(ev)*0.8, vmax=max(ev))
cmap = cm.ScalarMappable(norm=norm, cmap=cm.hot)
ec = [cmap.to_rgba(cl) for cl in ev]

# color the edges in the original graph with betweenness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='k', node_size=0, node_color='w', node_edgecolor='gray', node_zorder=2,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1, filepath=r"images/edge_centrality_betweenness.png", 
                        save=True, show=False, close=True, dpi=300)

# color the edges in the original graph with betweenness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='k', node_size=0,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1)
```

<b>Betweenness centrality</b> quantifies how often shortest paths pass through a node. These are considered as fastest (shortest) routes that connect different directions of Rockford city to the center of the city. The hotter lines have more betweenness that refers to more nodes are connected by shortest paths

```python
import time
start = time.time()

edge_centrality_degree = nx.degree_centrality(nx.line_graph(G))

end = time.time()
print(f'Elapsed time -> {round(float(end-start),2)} seconds')
```

```python
# list of edge values for the orginal graph
ev = [edge_centrality_degree[edge + (0,)] for edge in G.edges()]

# color scale converted to list of colors for graph edges
norm = colors.Normalize(vmin=min(ev)*0.8, vmax=max(ev))
cmap = cm.ScalarMappable(norm=norm, cmap=cm.hot)
ec = [cmap.to_rgba(cl) for cl in ev]

# color the edges in the original graph with betweenness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='k', node_size=0, node_color='w', node_edgecolor='gray', node_zorder=2,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1, filepath=r"images/edge_centrality_degree.png", 
                        save=True, show=False, close=True, dpi=300)

# color the edges in the original graph with betweenness centralities in the line graph
fig, ax = ox.plot_graph(G, bgcolor='k', node_size=0,
                        edge_color=ec, edge_linewidth=1.5, edge_alpha=1)
```

<b>Degree centrality</b> refers to number of nearest neighbors to a node. From the above graph, there is a high degree centrality within Rockford. Therefore, the interpretation is more nodes are connected allowing drivers to reroute better with short distances and divert the traffic flows into different routes without causing delays in reaching the desired destinations.

```python
type(edge_centrality_degree)
```

Degree centrality histogram is a <b>non-symmetric left-skewed distribution</b> suggesting that there more nodes with longer distances with higher frequency

```python
edge_centrality_values = list(edge_centrality_degree.values())
plt.hist(edge_centrality_values, bins = 10)
plt.savefig(r"images/edge_centrality_degree_hist.png")
plt.title('Degree centrality of Rockford')
plt.show()
```

Closeness centrality histogram is a <b>non-symmetric left-skewed distribution</b> suggesting that there more nodes with longer distances with higher frequency

```python
edge_closeness_values = list(edge_centrality_closeness.values())
plt.hist(edge_closeness_values, bins = 10)
plt.savefig(r'images/edge_closeness_values_hist.png')
plt.title('Closeness centrality of Rockford')
plt.show()
```

Betweeness histogram is a <b>right-skewed distribution</b> suggesting that there more nodes with shorter distances with higher frequency

```python
edge_betweenness_values = list(edge_centrality_betweenness.values())
plt.hist(edge_betweenness_values, bins = 10)
plt.savefig(r"images/edge_betweenness_values_hist.png")
plt.title('Betweenness centrality of Rockford')
plt.show()
```

Calculate the <b> shortest path </b> between two nodes and visualize it. Consider two random nodes from any of the computation metrics available (degree/closeness/betweenness) and calculate the shortest path using networkx module

```python
origin = 237525715 
destination = 10657296055
```

```python
# calculates the shortest path between two nodes along the network using Djikstra algorithm
route = nx.shortest_path(G, origin, destination)
# visualizes network and overlays the shortest path
fig, ax = ox.plot_graph_route(G, route, route_color='green', route_linewidth=6,
                              node_color='#2c7fb8', node_size = 2, bgcolor='white')
```

```python
# Calculate the shortest path length between the origin and destination nodes
length = nx.shortest_path_length(G, origin, destination, weight='length')

# Print the shortest path length
print(f"The shortest path length is {round(length*0.000621371,2)} miles.")
```
